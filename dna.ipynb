```python
def main():
    infile = open('human_insulin.fas', 'r')
    line1 = infile.readline()
    sequence = infile.read()
    total_num = no_of_sequence(sequence)
    print("This is a", dna_or_rna(sequence), "sequence")
    print("                                                         ")
    print("The total number of nucleobases found in the DNA sequence in the file is ",total_num)
    print("                                                         ")
    comp__percent(sequence, total_num)
    print("                                                         ")
    reverse_compliment(sequence)
    print("                                                         ")
    m_rna_sequence = m_rna(sequence)
    print("mRNA Sequence:", m_rna(sequence))
    print("                                                         ")
    p = protein_sequence(m_rna_sequence)
    print("Protein sequence:", p)
    print("                                                         ")
    _percentage_amino_acid(p)

    infile.close()


def no_of_sequence(sequence):
    counter = 0
    for n in sequence:
        if n == 'A' or n == 'C' or n == "G" or n == "T":
            counter += 1
    return counter


def comp__percent(sequence, total_num):
    a_counter = 0
    c_counter = 0
    t_counter = 0
    g_counter = 0
    for n in sequence:
        if n == 'A':
            a_counter += 1
        elif n == 'C':
            c_counter += 1
        elif n == 'T':
            t_counter += 1
        elif n == 'G':
            g_counter += 1
    comp__percent_adenine = (a_counter/total_num)*100
    comp__percent_cytosine = (c_counter / total_num) * 100
    comp__percent_thymine = (t_counter / total_num) * 100
    comp__percent_guanine = (g_counter / total_num) * 100
    print("The nucleotide composition in _percentage for Adenine is ", comp__percent_adenine, "%\n",
          "The nucleotide composition in _percentage for Cytosine is ", comp__percent_cytosine, "%\n",
          "The nucleotide composition in _percentage for Thymine is ", comp__percent_thymine, "%\n",
          "The nucleotide composition in _percentage for Guanine is ", comp__percent_guanine, "%")


def dna_or_rna(sequence):
    for n in sequence:
        if n == 'T':
            return "DNA"
        elif n == "U":
            return "RNA"


def reverse_compliment(sequence):
    compliment = {'A': 'T', 'C': 'G', 'T': 'A', 'G': 'C'}
    compliment_sequence_list = []
    sequence_list= list(sequence)
    for n in sequence_list:
        for key in compliment:
            if n == key:
                compliment_sequence_list+=compliment[key]

    print("Reverse compliment: ","".join(compliment_sequence_list[::-1]))

def m_rna(sequence):
    new_sequence= sequence.replace('T', 'U')
    return new_sequence


def protein_sequence(m_rna_sequence):
    protein = proteins = {'UUU': 'F', 'UUC': 'F',
                'UUA': 'L', 'UUG': 'L', 'CUU': 'L', 'CUC': 'L', 'CUA': 'L', 'CUG': 'L',
                'AUU': 'I', 'AUC': 'I', 'AUA': 'I',
                'AUG': 'M', 'GUU': 'V', 'GUC': 'V', 'GUA': 'V', 'GUG': 'V',
                'UCU': 'S', 'UCC': 'S', 'UCA': 'S', 'UCG': 'S',
                'CCU': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P',
                'ACU': 'T', 'ACC': 'T', 'ACA': 'T', 'ACG': 'T',
                'GCU': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A',
                'UAU': 'Y', 'UAC': 'Y', 'UAA': 'STOP', 'UAG': 'STOP',
                'CAU': 'H', 'CAC': 'H', 'CAA': 'Q', 'CAG': 'Q',
                'AAU': 'N', 'AAC': 'N', 'AAA': 'K', 'AAG': 'K',
                'GAU': 'D', 'GAC': 'D', 'GAA': 'E', 'GAG': 'E',
                'UGU': 'C', 'UGC': 'C', 'UGA': 'STOP',
                'UGG': 'W', 'CGU': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R',
                'AGU': 'S', 'AGC': 'S', 'AGA': 'R', 'AGG': 'R',
                'GGU': 'G', 'GGC': 'G', 'GGA': 'G', 'GGG': 'G'}
    m_rna_sequence_stripped = ""

    for n in m_rna_sequence:
        if n == 'A' or n == 'C' or n == "G" or n == "U":
            m_rna_sequence_stripped+=n

    group = [m_rna_sequence_stripped[i:i+3] for i in range(len(m_rna_sequence_stripped)-2)]


    ps_group = []
    for x in group:
        for value in protein:
            if x == value:
                ps_group += protein[value]


    return ps_group

def _percentage_amino_acid(protein_sequence):
    total_num = len(protein_sequence)
    g_counter = 0
    r_counter =0
    l_counter =0
    f_counter =0
    v_counter = 0
    s_counter = 0
    p_counter = 0
    t_counter = 0
    a_counter = 0
    y_counter = 0
    h_counter = 0
    n_counter = 0
    q_counter = 0
    k_counter = 0
    d_counter = 0
    c_counter = 0
    w_counter = 0
    i_counter = 0
    m_counter = 0
    e_counter = 0
    
    for char in protein_sequence:
        if (char == 'G'):
            g_counter += 1
        elif (char == 'R'):
            r_counter +=1
        elif(char == 'L'):
            l_counter += 1
        elif(char == 'F'):
            f_counter += 1
        elif(char == 'V'):
            v_counter += 1
        elif(char == 'S'):
            s_counter += 1
        elif(char == 'P'):
            p_counter += 1
        elif(char == 'T'):
            t_counter += 1
        elif(char == 'A'):
            a_counter += 1
        elif(char == 'Y'):
            y_counter += 1
        elif(char == 'H'):
            h_counter += 1
        elif(char == 'N'):
            n_counter += 1
        elif(char == 'Q'):
            q_counter += 1
        elif(char == 'K'):
            k_counter += 1
        elif(char == 'D'):
            d_counter += 1
        elif(char == 'C'):
            c_counter += 1
        elif(char == 'W'):
            w_counter += 1
        elif(char == 'I'):
            i_counter += 1
        elif(char == 'M'):
            m_counter += 1
        elif(char == 'E'):
            e_counter += 1

    e_percent = (e_counter/total_num) * 100
    m_percent = (m_counter/total_num) * 100
    i_percent = (i_counter/total_num) * 100
    w_percent = (w_counter/total_num) * 100
    c_percent = (c_counter/total_num) * 100
    d_percent = (d_counter/total_num) * 100
    k_percent = (k_counter/total_num) * 100
    q_percent = (q_counter/total_num) * 100
    n_percent = (n_counter/total_num) * 100
    h_percent = (h_counter/total_num) * 100
    y_percent = (y_counter/total_num) * 100
    a_percent = (a_counter/total_num) * 100
    t_percent = (t_counter/total_num) * 100
    p_percent = (p_counter/total_num) * 100
    s_percent = (s_counter/total_num) * 100
    v_percent = (v_counter/total_num) * 100
    f_percent = (f_counter/total_num) * 100
    l_percent = (l_counter/total_num) * 100
    r_percent = (r_counter/total_num) * 100
    g_percent = (g_counter/total_num) * 100
    
    print("Percentage amino acid composition:\n",
                "Arginine (R): ", r_percent, "\n",
                "Glycine (G): ", g_percent, "\n",
                "Glutamate (E): ",e_percent, "\n",
                "Methionine (M): ",m_percent, "\n",
                "Isoleucine (I): ", i_percent, "\n",
                "Tryptophan (W): ", w_percent, "\n",
                "Cysteine (C): ", c_percent,"\n",
                "Aspartate (D): ", d_percent, "\n",
                "Lysine (K): ", k_percent,"\n",
                "Glutamine (Q): ", q_percent,"\n",
                "Asparagine (N): ", n_percent,"\n",
                "Histidine (H): ",h_percent,"\n",
                "Tyrosine (Y): ", y_percent,"\n",
                "Alanine (A): ",a_percent,"\n",
                "Threonine (T): ", t_percent,"\n",
                "Proline (P): ",p_percent,"\n",
                "Serine (S): ", s_percent,"\n",
                "Valine (V): ",v_percent,"\n",
                "Phenylalnine (F): ", f_percent, "\n",
                "Leucine (L): ", l_percent,"\n")
    
    protein_percent = {"Arginine (R)":r_percent,
                "Glycine (G) ": g_percent,
                "Glutamate (E)": e_percent, 
                "Methionine (M)":m_percent,
                "Isoleucine (I)": i_percent, 
                "Tryptophan (W)": w_percent, 
                "Cysteine (C)": c_percent, 
                "Aspartate (D)": d_percent, 
                "Lysine (K)": k_percent, 
                "Glutamine (Q)":q_percent, 
                "Asparagine (N)":n_percent, 
                "Histidine (H)": h_percent, 
                "Tyrosine (Y)": y_percent, 
                "Alanine (A)":a_percent, 
                "Threonine (T)": t_percent, 
                "Proline (P)":p_percent, 
                "Serine (S)":s_percent, 
                "Valine (V)":v_percent, 
                "Phenylalnine (F)": f_percent,  
                "Leucine (L)":l_percent }
    
    fig, ax =plt.subplots(figsize = (50,50))
    ax.bar(protein_percent.keys(), protein_percent.values())
    plt.xticks(rotation = 45)
    ax.set(title = "Percentage amino acid composition",
    ylabel = "Percentage composition");
    fig.savefig("images/pCompostionBarChart.png")
    
main()
```
